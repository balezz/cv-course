## 1. Классический CV, часть 1  

#### Популярные фреймворки - numpy, opencv, matplotlib, albumentations  

CV основано на применении различных алгоритмов к изображениям, представленных в виде N-мерных матриц. Для численных вычислений в python самой популярной библиотекой является numpy. В нем реализованы эффективные способы представления изображений в различных типах данных (uint, int, float, complex). Также в нем можно найти множество алгоритмов над многомерными массивами: арифметические и логические операции, изменение размерности, сортировка и поиск, ввод-вывод, преобразования Фурье, линейная алгебра, статистика и многое другое. Тесную интеграцию с numpy имеют библиотеки opencv scikit-image matplotlib

Создадим единичную матрицу и отобразим ее с помощью matplotlib 

```python
import numpy as np
eye = np.eye(10)
eye 
```

```python
import matplotlib.pyplot as plt
plt.imshow(eye, cmap='gray')
```

Большие возможности по обработке изображений предоставляет библиотека opencv. Она тесно интегрирована с numpy и при чтении изображения из файла представляет его в виде массива ndarray.

```python
import cv2
img = cv2.imread('img/vitara.jpg')
type(img), img.shape
```

Из-за разного следования BGR в opencv и RGB в matplotlib при отображении может нарушиться цветопередача. 

```python
plt.imshow(img)
```

Один из простых способов переставить цвета местами - с помощью индексации массивов. Исправим это, заодно увеличив размер отображения.

```python
plt.figure(figsize=(20,14))
plt.imshow(img[:,:, ::-1])
```

Можно использовать встроенный механизм отрисовки изображений IPython. Его особенность состоит в том, что он требует кодирования матрицы обратно в какой-либо формат, например JPEG. Для удобства можно обвернуть две строчки кода в функцию.

```python
from IPython.display import Image, display
def show(img):
    _, buffer = cv2.imencode('.jpg', img)
    display(Image(data=buffer))
show(img)
```
